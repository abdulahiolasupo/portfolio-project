#### How to Create a Scatterplot in Python Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
day = [1,2,3,4,5,6,7]
number = [3,4,6,2,1,7,5]


n = ['yellow','blue','green','purple','k','pink','orange']
s = [100,150,200,250,300,350,400]


plt.scatter(day, number, color= n, sizes=s, marker='*', edgecolor='red', linewidth=1)


plt.title('Scatter Plot', fontsize=15)
plt.xlabel('day', fontsize=15)
plt.ylabel('number', fontsize=15)

plt.show()
```

```python
day = [1,2,3,4,5,6,7]
number = [3,4,6,2,1,7,5]


n = ['yellow','blue','green','purple','k','red','orange']
s = [100,150,200,250,300,350,400]


plt.scatter(day, number, color= n, sizes=s, alpha=0.4, marker='*')


plt.title('Scatter Plot', fontsize=15)
plt.xlabel('day', fontsize=15)
plt.ylabel('number', fontsize=15)

plt.show()
```

```python
day = [1,2,3,4,5,6,7]
number = [3,4,6,2,1,7,5]
number1 = [6,3,5,1,4,8,2]

color = [10,20,30,40,50,60,70]
color1 = [25,35,30,45,55,65,75]
## color map(cmap) function.  check color map website


sizes = [100,150,200,250,300,350,400]


plt.scatter(day, number, c= color, s=sizes, cmap='viridis', alpha=0.4)
plt.scatter(day, number1, c=color1, cmap='BrBG', s=sizes)


plt.title('Scatter Plot', fontsize=15)
plt.xlabel('day', fontsize=15)
plt.ylabel('number', fontsize=15)
t = plt.colorbar()
t.set_label('ColorBar', fontsize=15)



plt.show()
```

#### Areas VS Stack plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = ['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l, colors=['r','g','m'])


plt.title('Python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc=2)
plt.show()
```
